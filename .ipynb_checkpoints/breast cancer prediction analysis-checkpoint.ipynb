```python
numerical_features=[feature for feature in df.columns if df[feature].dtype!='O']
numerical_features
```

```python
categorical_feature=[feature for feature in df.columns if feature not in numerical_features]
categorical_feature
```

```python
# Checking the cardinality of the numerical features and printing features whose cardinality is less than 20 
import numpy as np
for feature in numerical_features:
    cardinal=len(np.unique(df[feature]))
    if cardinal<=20:
            print(f"The cardinality of {feature} is {cardinal} ")
```

```python
# There is no cardinality in the numerical features
import seaborn as sns
import matplotlib.pyplot as plt
for feature in numerical_features:
    sns.displot(df[feature],kind='kde')
    plt.title(feature)
    plt.show()
```

```python
# The most of the distribution of the feature is slightly skewwed 
# Let's try qqplot
import statsmodels.api as sm
for feature in numerical_features:
    sm.qqplot(df[feature],line='s')
    plt.title(feature)
    plt.show()
```

```python
# Finding the features having 0 in it
data=df.copy()
for feature in numerical_features:
    count=0
    for i in data[feature]:
        if i==0:
            count +=1
    if count!=0:
        print(f"{feature} has {count} 0\'s")
```

```python
feature_with_zero=[feature for feature in numerical_features if 0 in data[feature].unique()]
feature_with_zero
```

```python
z=df[feature_with_zero]
z[z==0].dropna(axis=0)
```
