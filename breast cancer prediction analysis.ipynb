```python
# Finding the outliers before log transformation
# By Z score
def Z_score(feature,data):
    outlier=[]
    mean=np.mean(data[feature])
    std=np.std(data[feature])
    for i in data[feature]:
        score=(i-mean)/std
        if np.abs(score)>3:
            outlier.append(i)
    return outlier

for feature in numerical_features:
    out=Z_score(feature,data)
    print(f"Outliers in {feature} : {len(out)} ")
```

```python
# By Inter quartile range
def IQR(feature,data):
    outlier=[]
    q1,q3=np.percentile(data[feature],[25,75])
    iqr=q3-q1
    lower=q1-1.5*(iqr)
    upper=q3+1.5*(iqr)
    for i in data[feature]:
        if i<lower or i>upper:
            outlier.append(i)
    return outlier
for feature in numerical_features:
    out=IQR(feature,data)
    print(f"Outliers in {feature} : {len(out)}")
```

```python
# log transformation
for feature in numerical_features:
    sm.qqplot(np.log(data[feature]),line='s')
    plt.title(f"{feature} after log transformation")
    plt.show()
```

```python
# Checking outliers after transformation by Z score
log_data=data.copy()
for feature in numerical_features:
    each=np.log(log_data[feature])
    outlier=[]
    mn=np.mean(each)
    st=np.std(each)
    for i in each:
        if np.abs((i -mn)/st) >3:
            outlier.append(i)
    print(f"Outliers of {feature} : {len(outlier)}")
```

```python
# Checking outliers after transformation by IQR range
for feature in numerical_features:
    each=np.log(log_data[feature])
    outlier=[]
    q1,q3=np.percentile(each,[25,75])
    iqr=q3-q1
    lower=q1-1.5*(iqr)
    upper=q3+1.5*(iqr)
    for i in each:
        if i<lower or i>upper:
            outlier.append(i)
    print(f"Outliers of {feature} : {len(outlier)}")
```

```python
data_1=np.log(data[numerical_features])
diagnosis_dumm=pd.get_dummies(data['diagnosis'],drop_first=True)
final_df=pd.concat([diagnosis_dumm,data_1],axis=1)
final_df.head()
```

```python
# Correlation
sns.set(rc={'figure.figsize':(24.7,15.27)})
sns.heatmap(final_df.corr(),annot=True)
```

```python
# sns.set(rc={'figure.figsize':(7.7,4.27)})
X=final_df.drop('M',axis=1)
y=final_df['M']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
# Using filter method for feature selection
# Finding the correlation of each independent feature with the dependent feature
for feature in X.columns:
    corr_val=np.abs(pd.concat([y,X[feature]],axis=1).corr().loc['M',feature])
    if corr_val>=0.65:
        print(f"{feature} has a high correlation with 'M' ")
```

```python
# Finding the high correlation among the independent features
lst=list(X.columns)
for  i in X.columns:
    for j in lst:
        if i!=j:
            corr_val=np.abs(pd.concat([X[i].to_frame().rename(columns={0:i}),X[j].to_frame().rename(columns={0:j})],axis=1).corr().loc[i,j])
            if corr_val>=0.85:
                print(f"{i} with {j}")
    lst.remove(i) 
```

```python
# Performing mutual information
from sklearn.feature_selection import mutual_info_classif
mi=mutual_info_classif(X_train,y_train)
mi
```

```python
grp=pd.Series(mi)
grp.index=X.columns
c=grp.sort_values(ascending=True)
```

```python
sns.set(rc={'figure.figsize':(20.7,9.27)})
c.plot.bar()
plt.show()
```

```python
grp.nlargest(n=20,keep='first')# keeping the top 20 features important features
```

```python
top_features1=grp.nlargest(n=20,keep='first').index
top_features1
```

```python
# Performing ANOVA test
from sklearn.feature_selection import f_classif
sel=f_classif(X_train,y_train)
sel
```

```python
p_values=pd.Series(sel[1])
p_values.index=X.columns
p_values.sort_values(ascending =True,inplace=True)
p_values.plot.bar()
```

```python
# Features whose p value is less than 0.05 are the important features
p_values[p_values<=0.05]
```

```python
imp_features2=list(p_values[p_values<=0.05].index)
```

```python
imp_features2
```

```python
#Getting the common features from the 2 tests
common_features=[]
for feature in imp_features2:
    if feature in top_features1:
        common_features.append(feature)
common_features
```

```python
# 'area_mean','radius_mean','area_worst','perimeter_se','radius_se' are highly correlated with each other
# Removing those features having very high correlation with each other
high_corr=['area_mean','radius_mean','area_worst','perimeter_se','radius_se']
common_features=[feature for feature in common_features if feature not in high_corr]
common_features
```

```python
# Performing feature Scaling
X_train=X_train[common_features]
X_test=X_test[common_features]
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
X_train_scale=mms.fit_transform(X_train)
X_train_scale
```

```python
X_train_scale=pd.DataFrame(X_train_scale,columns=common_features)
X_train_scale
```

```python
X_test_scale=mms.transform(X_test)
X_test_scale=pd.DataFrame(X_test_scale,columns=common_features)
```

```python
# Doing crossvalidation on multiple models to get the best model
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xg
```

```python
from sklearn.metrics import accuracy_score
lr=LogisticRegression()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
gb=GradientBoostingClassifier()
ada=AdaBoostClassifier()
svc=SVC()
knn=KNeighborsClassifier()
xgb=xg.XGBClassifier()
scores=[[],[],[],[],[],[],[],[]]
classifiers=[lr,dt,rf,gb,ada,svc,knn,xgb]
kf = KFold(n_splits=5)
for train_index, test_index in kf.split(X_train_scale):
    X_train_fold, X_test_fold = X_train_scale.iloc[train_index], X_train_scale.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]
    for index,model in enumerate (classifiers):
        model.fit(X_train_fold,y_train_fold)
        p=model.predict(X_test_fold)
        acc=accuracy_score(y_test_fold,p)
        scores[index].append(acc)
```

```python
for index,each in enumerate(scores):
    print(f"KFold accuracies of {classifiers[index]} : {each}")
```

```python
# Geting mean of each models
for index,each in enumerate(scores):
    print(f"Mean of KFold accuracies of {classifiers[index]} : {np.mean(each)*100}")
    
```
